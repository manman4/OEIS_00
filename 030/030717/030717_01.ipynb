```python
from itertools import chain

def uniq(li):
    li_uniq = []
    for x in li:
        if x not in li_uniq:
            li_uniq.append(x)
    return li_uniq

N=30
A030717_list = [1]
for _ in range(1, N+1):
    A030717_list += [A030717_list.count(i) for i in sorted(uniq(A030717_list))]

print(A030717_list)
```

```python
import matplotlib.pyplot as plt

# 1-
x = list(range(1, len(A030717_list)+1))
plt.plot(x, A030717_list)
```
